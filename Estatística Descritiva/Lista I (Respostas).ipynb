## Questão 01

```python
# Importando dataset da Tabela 2.1: Informações sobre estado civil, grau de instrução,
# número de filhos, salário, idade e procedência de 36 empregados da seção de orçamentos da Companhia MB
dataset = pd.read_csv('data/tabela2_1.csv', encoding = 'latin1')

dataset['Idade'] = [(int(x) + int(y)*0.1) for x,y in zip(dataset['Anos'], dataset['Meses'])]
dataset['N de Filhos'].fillna(0, inplace = True)

dataset.head(5)
```

A tabela acima importada possui as seguintes variáveis:

<ul>
    <li><b>Estado civil</b>: é uma <i>variável qualitativa nominal</i>, pois assume somente valores que são rótulos e não possuem uma ordenação natural; a escala adotada é a nominal, sendo assim não podemos realizar operações aritméticas com ela; uma medida de posição apropriada é a moda e a variável pode ser representada visualmente através de um gráfico de setores (gráfico de pizza) ou de um gráfico de barras, mas o <i>gráfico de pizza é preferível</i> já que não há nenhum tipo de hierarquia entre os valores assumidos;</li>
    <li><b>Grau de instrução</b>: é uma <i>variável qualitativa ordinal</i>, pois assume rótulos que indicam ordenação crescente relacionada, neste caso, com o tempo de estudo; a escala ordinal é adotada aqui, de forma que um valor é diferente e maior/menor que outro; medidas de posição apropriadas são a mediana e a moda; novamente o gráfico de pizza e o gráfico de barras podem ser usados, mas o de barras é preferível por apresentar os resultados de forma ordenada;</li>
    <li><b>Número de filhos</b>: é uma <i>variável quantitativa discreta</i>, pois assume valores que são resultado de contagens; a escala racional é adotada aqui, pois existe um zero absoluto (ausência de filhos) e pode-se estabelecer uma proporção entre os dados, dois a dois; aqui podemos fazer operações aritméticas, e média, mediana e moda são medidas de posição apropriadas; um <i>gráfico de barras</i> é adequada para representar visualmente o conjunto de dados, devido à ordenação implícita que pode ser estabelecida;</li>
    <li><b>Salário</b>: é uma <i>variável quantitativa contínua</i>, pois assume valores no intervalo dos números reais, e a escala racional também é adotada aqui; um <i>histograma</i> é mais adequado para representar o conjunto de dados, já que o agrupamento dos dados em classes permite uma melhor visualização;</li>
    <li><b>Idade</b>: esta variável está dividida em duas variáveis quantitativas discretas (anos e meses) e a partir delas podemos estabelecer uma <i>variável quantitativa contínua</i> definida pela fórmula $idade = anos + meses \times 0,1$, que mapeia valores no intervalo dos números reais; sendo assim, à semelhança da variável <i>Salário</i>, o <i>histograma</i> é preferível para representá-la pelas mesmas razões descritas anteriormente;</li>
    <li><b>Região de procedência</b>: é uma variável qualitativa normal à semelhança de <i>Estado civil</i> e o <i>gráfico de pizza</i> é preferível pelas mesmas razões.</li>
</ul>

### Estado civil

```python
s_est_civil = sum(dataset['Estado Civil'].value_counts())
freq_est_civil = [(x/s_est_civil)*100 for x in dataset['Estado Civil'].value_counts()]
dataset['Estado Civil'].value_counts(), freq_est_civil
```

```python
est_civil_labels = [x.capitalize() for x in set(dataset['Estado Civil'])]

plt.pie(freq_est_civil, labels = est_civil_labels, autopct='%1.1f%%', shadow = True)
plt.title('Estado civil dos empregados da Companhia MB')
plt.show()
```

### Grau de instrução

```python
s_grau = sum(dataset['Grau de Instrução'].value_counts())
freq_grau = [(x/s_grau)*100 for x in dataset['Grau de Instrução'].value_counts()]
dataset['Grau de Instrução'].value_counts(), freq_grau
```

```python
grau_labels = ['Ensino Fundamental', 'Ensino Médio', 'Superior']
freq_grau = [33.3, 50.0, 16.6]
y_pos = np.arange(len(grau_labels))
 
plt.bar(y_pos, freq_grau, align = 'center', alpha = 0.5)
plt.xticks(y_pos, grau_labels)
plt.ylabel('Porcentagem (%)')
plt.title('Grau de instrução dos empregados da Companhia MB')
plt.show()
```

### Número de filhos

```python
s_filhos = sum(dataset['N de Filhos'].value_counts())
freq_filhos = [(x/s_filhos)*100 for x in dataset['N de Filhos'].value_counts()]
dataset['N de Filhos'].value_counts(), freq_filhos
```

```python
filhos_labels = ['0', '1', '2', '3', '4', '5']
freq_filhos = [55.6, 13.9, 19.4, 8.3, 0.0, 2.8]
y_pos = np.arange(len(filhos_labels))
 
plt.bar(y_pos, freq_filhos, align = 'center', alpha = 0.5)
plt.xticks(y_pos, filhos_labels)
plt.ylabel('Porcentagem (%)')
plt.xlabel('Quantidade de filhos')
plt.title('Números de filhos dos empregados da Companhia MB')
plt.show()
```

### Salário

```python
sal_ord = sorted([float(x.replace(',', '.')) for x in dataset['Salario (x Sal Min)']])
amplitude = sal_ord[-1] - sal_ord[0]
t5 = amplitude/5
t8 = amplitude/8

print(sal_ord[0], sal_ord[-1])
print(t8,t5)

# Amplitude das classes = 3
```

```python
bins = [4, 7, 10, 13, 16, 19, 22, 25]

plt.hist(sal_ord, bins = bins, edgecolor = 'black', linewidth = 1.2)
plt.title('Salários dos empregados da Companhia MB')
plt.xlabel('Salário')
plt.ylabel('Frequência')

plt.show()
```

### Idade

```python
idade_ord = sorted(dataset['Idade'])
amplitude = idade_ord[-1] - idade_ord[0]
t5 = amplitude/5
t8 = amplitude/8

print(idade_ord[0], idade_ord[-1])
print(t8,t5)

# Amplitude das classes = 4
```

```python
bins = [21, 26, 31, 36, 41, 46, 51]

plt.hist(idade_ord, bins = bins, edgecolor = 'black', linewidth = 1.2)
plt.title('Idades dos empregados da Companhia MB')
plt.xlabel('Idade')
plt.ylabel('Frequência')

plt.show()
```

### Região de procedência

```python
s_regiao = sum(dataset['Região de Procedência'].value_counts())
freq_regiao = [(x/s_regiao)*100 for x in dataset['Região de Procedência'].value_counts()]
dataset['Região de Procedência'].value_counts(), freq_regiao
```

```python
regiao_labels = [x.capitalize() for x in set(dataset['Região de Procedência'])]

plt.pie(freq_regiao, labels = regiao_labels, autopct='%1.1f%%', shadow = True)
plt.title('Região de procedência dos empregados da Companhia MB')
plt.show()
```
